```python
import ast
```

```python
def convert(text):
    # Check if the input is already a list
    if isinstance(text, list):
        return text  # If it is, return it directly
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
movies['genres']=movies['genres'].apply(convert)
# movies.head()
```

```python
movies.dropna(inplace=True)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
def convert3(text):
    # Check if the input is already a list of strings
    if isinstance(text, list) and all(isinstance(item, str) for item in text):
        return text  # If it is, return it directly
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter != 3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
movies.head()
```

```python
movies.head()
```

```python
def fetch_director(text):
    # Check if the input is already a list of strings
    if isinstance(text, list) and all(isinstance(item, str) for item in text):
        return text  # If it is, return it directly
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
def collapse(L):
    L1 = []
    for i in L:
        #Only collapse multi-word strings to avoid removing spaces within words
        if " " in i:
            L1.append(i.replace(" ",""))
        else:
            L1.append(i)
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
def recommend(movie):
    index = new_df[new_df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title)

```
